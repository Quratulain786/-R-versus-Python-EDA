```python
# Encode categorical variables
label_encoders = {}
for column in ['team1', 'team2', 'venue', 'winner']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
# Print the mapping of encoded values to original country names for each column
for column in ['team1', 'team2', 'venue', 'winner']:
    print(f"Column: {column}")
    print("Encoded Value : Original Country")
    for code, value in zip(range(len(label_encoders[column].classes_)), label_encoders[column].classes_):
        print(f"    {code} : {value}")
```

```python
# Decode the labels for better readability in the plot
data['team1'] = label_encoders['team1'].inverse_transform(data['team1'])
data['team2'] = label_encoders['team2'].inverse_transform(data['team2'])
data['winner'] = label_encoders['winner'].inverse_transform(data['winner'])
# Create a pivot table to count wins
pivot_table = data.pivot_table(index='winner', columns='team2', aggfunc='size', fill_value=0)

# Plot the grouped bar chart
pivot_table.plot(kind='bar', figsize=(14, 8), width=0.8)
plt.xlabel('Winner Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins by Each Country Against Each Opponent')
plt.xticks(rotation=90)
plt.legend(title='Opponent Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('ODI_Match_info.csv')


# Counting occurrences of each team in 'winner' column
winner_counts = df['winner'].value_counts()

# Selecting top 10 winners
top_winners = winner_counts.head(10)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
colors = plt.cm.autumn_r(range(len(top_winners)))
plt.barh(top_winners.index, top_winners.values, color=colors)
plt.xlabel('Number of Wins', fontsize=12)
plt.ylabel('Teams', fontsize=12)
plt.title('Top 10 High Performing Winners', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest number of wins at the top

# Adding data labels on bars
for index, value in enumerate(top_winners.values):
    plt.text(value + 0.1, index, str(value), va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('ODI_Match_info.csv')

# Counting occurrences of each country in 'winner' column
winner_counts = df['winner'].value_counts()

# Selecting bottom 10 performers (countries with fewest wins)
bottom_performers = winner_counts.tail(17)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
colors = plt.cm.winter_r(range(len(bottom_performers)))

plt.barh(bottom_performers.index, bottom_performers.values, color=colors)
plt.xlabel('Number of Wins', fontsize=12)
plt.ylabel('Countries', fontsize=12)
plt.title('Top 10 Poor Performing Countries', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to have the lowest number of wins at the top

# Adding data labels on bars
for index, value in enumerate(bottom_performers.values):
    plt.text(value + 0.1, index, str(value), va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Counting wins and losses for each country
country_stats = df.apply(lambda x: pd.Series([x['team1'], x['team2'], 'win' if x['team1'] == x['winner'] else 'loss'], index=['country', 'opponent', 'result']), axis=1)
country_stats = pd.crosstab(country_stats['country'], country_stats['result'])

# Selecting top 6 countries based on number of wins
top_countries = country_stats['win'].nlargest(6)

# Plotting multiple bar charts for top 5 countries with adjusted y-axis
plt.figure(figsize=(12, 8))

for i, country in enumerate(top_countries.index):
    ax = plt.subplot(2, 3, i+1)  # Adjust subplot layout as per the number of top countries
    bars = ax.bar(country_stats.columns, country_stats.loc[country], color=['pink', 'green'])
    ax.set_title(f'{country} - Win/Loss Record')
    ax.set_ylabel('Number of Matches')

    # Calculate y-axis limit based on max value in current country's data
    max_value = country_stats.loc[country].max()
    ax.set_ylim(0, max_value + 15)  # Set y-axis limit just above the maximum value

    # Add data labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.2, yval, ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
